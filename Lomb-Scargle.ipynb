# Lomb-Scargle

* The Lomb-Scargle Periodogram 
 * after Lomb (1976) and Scargle (1982) 
* statistical tool designed to detect periodic signals in unevenly-spaced observations
* least squares fit of sinusoids to data samples
* also called Least-squares spectral analysis (LSSA)


```python
fig, axs = plt.subplots(nrows=2, ncols=1, num=None, figsize=(15, 6), dpi=80, facecolor='w', edgecolor='k')

def sin_func(t,ph=0):
  return np.sin( (t/350.) * 2. * np.pi  + ph)*1. + 100.

T1 = np.linspace(0,2000,1000)
T2 = np.linspace(0,2000,1000)
T3 = np.linspace(0,2000,1000)
Y1 = sin_func(T1,np.pi/16.)
Y2 = sin_func(T2, 0)
Y3 = sin_func(T3,-np.pi/16.)

f = np.logspace(np.log10(1./500), np.log10(1./100.), 1000)


axs[0].plot(T1, Y1, lw=0, marker=".", ms=1.0, label=str(len(T1)))
axs[0].plot(T2, Y2, lw=0, marker=".", ms=1.0, label=str(len(T2)))
axs[0].plot(T3, Y3, lw=0, marker=".", ms=1.0, label=str(len(T3)))
# axs[0].set_xlim([np.min(T1),np.max(T1)])

axs[1].plot(1./f, np.sqrt(4.*( scipy.signal.lombscargle(T1, (Y1)-np.mean(Y1), f*2.*np.pi) / T1.shape[0] )), lw=0, marker='.')
axs[1].plot(1./f, np.sqrt(4.*( scipy.signal.lombscargle(T2, (Y2)-np.mean(Y2), f*2.*np.pi) / T2.shape[0] )), lw=0, marker='.')
axs[1].plot(1./f, np.sqrt(4.*( scipy.signal.lombscargle(T3, (Y3)-np.mean(Y3), f*2.*np.pi) / T3.shape[0] )), lw=0, marker='.')
# axs[1].plot(f, npgram)
# axs[1].set_xscale("log")
axs[1].set_xlim([200,500])
axs[1].set_xlabel("Period in d")

""
```

# Read LFBD, LFST, LFSR

```python
EBFS = pd.read_pickle("../muon/EBFS.pkl")
print EBFS.info(max_cols=2)
```

```python
EBFS.head(3)
```

```python
import pandas as pd
LFBD = pd.read_csv("LFBD.txt", sep='\t', comment="#", header=1, usecols=[1,2], names=['date','tmpc'], index_col=False, na_values=['M'], dtype={'date':'str', 'tmpc':'float32'} )
LFBD = LFBD[~pd.isnull(LFBD['tmpc'])]
LFBD['Start'] = pd.to_datetime(LFBD['date']).astype(np.int64) // 10**9 # convert to unix timestamp
LFBD['Temp'] = LFBD['tmpc'].astype(np.float32)+273.15
LFBD.drop([ 'date','tmpc' ], axis=1, inplace=True)
LFBD.set_index('Start', inplace=True)
print LFBD.info()
```

```python
LFBD.head(3)
```

```python
import pandas as pd
LFST = pd.read_csv("LFST.txt", sep='\t', comment="#", header=1, usecols=[1,2], names=['date','tmpc'], index_col=False, na_values=['M'], dtype={'date':'str', 'tmpc':'float32'} )
LFST = LFST[~pd.isnull(LFST['tmpc'])]
LFST['Start'] = pd.to_datetime(LFST['date']).astype(np.int64) // 10**9 # convert to unix timestamp
LFST['Temp'] = LFST['tmpc'].astype(np.float32)+273.15
LFST.drop([ 'date','tmpc' ], axis=1, inplace=True)
LFST.set_index('Start', inplace=True)
print LFST.info()
```

```python
LFST.head(3)
```

```python
import pandas as pd
LFSR = pd.read_csv("LFSR.txt", sep='\t', comment="#", header=1, usecols=[1,2], names=['date','tmpc'], index_col=False, na_values=['M'], dtype={'date':'str', 'tmpc':'float32'} )
LFSR = LFSR[~pd.isnull(LFSR['tmpc'])]
LFSR['Start'] = pd.to_datetime(LFSR['date']).astype(np.int64) // 10**9 # convert to unix timestamp
LFSR['Temp'] = LFSR['tmpc'].astype(np.float32)+273.15
LFSR.drop([ 'date','tmpc' ], axis=1, inplace=True)
LFSR.set_index('Start', inplace=True)
print LFSR.info()
```

# EBFS vs. LFBD

```python
import scipy
from scipy import interpolate

main_x = EBFS.index # x-sampling from here
main_y = EBFS['Temp']
comp_x = LFBD.index # will be resampled
comp_y = LFBD['Temp']

# select only common x values
main_x = main_x[( main_x >= max(main_x.min(),comp_x.min()) ) & ( main_x <= min(main_x.max(),comp_x.max()) )]
main_y = main_y[main_x]
# main_ref_y = scipy.interpolate.interp1d(main_x, main_y, kind='linear', assume_sorted=True, fill_value=(main_y[0],main_y[-1]))(main_x+0.)

print main_x.shape
print main_y.shape
print comp_x.shape
print comp_y.shape
print comp_x.values[0:10]
print comp_y.values[0:10]
print comp_y.iloc[0],comp_y.iloc[-1]

comp_ref_y = scipy.interpolate.interp1d(comp_x.values, comp_y.values, kind='nearest', assume_sorted=True, fill_value=(comp_y.iloc[0],comp_y.iloc[-1]))(main_x)

print comp_ref_y.shape



```
